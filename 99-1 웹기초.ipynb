## 웹기초
### 파이썬으로 웹페이지 접속하기

```python
import requests
from bs4 import BeautifulSoup

with requests.Session() as s:
    result = s.get("https://ko.wikipedia.org/wiki/피카츄")
    print(result.text)
```

```python
soup = BeautifulSoup(result.text, 'html.parser')
ps = soup.select(".mw-parser-output p")
print(ps[0].text)
```

```python
lists = ["피카츄","라이츄","파이리","꼬부기","버터플","야도란"]
```

```python
with requests.Session() as s:
    
    for list in lists:    
        result = s.get("https://ko.wikipedia.org/wiki/" + list)
        soup = BeautifulSoup(result.text, 'html.parser')
        ps = soup.select(".mw-parser-output p")
        
        try:
            print(list)
            print(ps[0].text)
            print("")
            print("")
        except:
            print(list)
            print("정보없음")
            print("")
            print("")
```
